The number of multiplication required to be performed is n * m * k, since the resulting matrix would be a n * m matrix and for each element k multiplactions are required to be performed.
The number of addition operations required to be performed is n * m * (k - 1), which is to add up all the products thus obtained.

```python
import time
import numpy as np
```

```python
tic = time.time()

A = [[1, 2, 3], [2, 3, 4]]
B = [[1, 2], [3, 4], [5, 6]]
res = [[0, 0], [0, 0]]

for i in range(len(A)):
    for j in range(len(B[0])):
        for k in range(len(B)):
            res[i][j] += A[i][k] * B[k][j]

print(res)

toc = time.time()
print(f"Total time taken: {toc - tic}")
```

```python
tic = time.time()

A = np.array([1, 2, 3, 2, 3, 4]).reshape(2, 3)
B = np.array([1, 2, 3, 4, 5, 6]).reshape(3, 2)

res = np.matmul(A, B)
print(res)

toc = time.time()
print(f"Total time taken: {toc - tic}")
```

The algorithm of finding the median using multiple passes will have a time complexity of O(n^2), since we'll be parsing through the entire array every time until we find the median value.

```python
A = [1, 4, 3, 2, 5, 6]

tic = time.time()

length = len(A)
median = -1
max = -1
max_2 = -1
for i in range(int(len(A)/2)+ 1):
    max_2 = max
    max = A[0]
    max_index = 0
    for j in range(len(A)):
        if(A[j] > max):
            max = A[j]
            max_index = j
    A.pop(max_index)

if(length % 2 == 0):
    print((max + max_2)/2)
else:
    print(max)

toc = time.time()
print(f"Total time taken: {toc - tic}")
```

```python
import jax
import jax.numpy as np

def my_function(x, y):
    return (x**2) * y + (y**3) * np.sin(x)

grad_fn_x = jax.grad(my_function, argnums=0)
grad_fn_y = jax.grad(my_function, argnums=1)

grad_x = grad_fn_x(0.0, 3.0)
grad_y = grad_fn_y(2.0, 0.0)

print(f"Gradient w.r.t to x: {grad_x}")
print(f"Gradient w.r.t y: {grad_y}")


```

```python
import matplotlib.pyplot as plt

x = np.arange(0.5, 100.0, 0.5)

def plot_function(y, title):
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(title)
    plt.plot(x, y)
    plt.show()


plot_function(x, '$y = x$')
plot_function(np.square(x), '$y = x^2$')
plot_function(np.power(x, 3)/100, '$y = x^3/100$')
plot_function(np.sin(x), '$y = sin(x)$')
plot_function(np.sin(x)/x, '$y = sin(x)/x$')
plot_function(np.log(x), '$y = log(x)$')
plot_function(np.exp(x), '$y = e^x$')
```
